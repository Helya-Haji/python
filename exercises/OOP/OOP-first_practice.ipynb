```python
class Student:
    def __init__(self, f_name, l_name, score, stu_num):
        self.f_name = f_name
        self.l_name = l_name
        self.score = score
        self.stu_num = stu_num

    def function_1(self , f_name):
        print (f"{f_name} is studing")

```

```python
S1 = Student("A" , "a" , 19 , "123A")
```

```python
students = []
for i in range(3):
    first_name = input("Enter the first name: ")
    last_name = input("Enter the last name: ")
    score = float(input("Enter the Score: "))
    num = input("Enter the student number: ")
    obj = Student (f_name = first_name, l_name = last_name , score = score , stu_num = num)
    students.append(obj)

print (students)
```
